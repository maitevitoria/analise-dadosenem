# Nota ENEM x IDHM

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mediamuni_df = pd.read_csv("MediaPorMunicipio.csv", sep=",", index_col=0)
```

```python
mediamuni_df.head()
```

```python
atlashumano_df =  pd.read_csv("atlas.csv", sep=",")
```

```python
pd.set_option('display.max_columns', 500)
```

```python
atlashumano_df = atlashumano_df[['ANO', 'UF', 'Codmun6' ,'Codmun7','Município', 'IDHM', 'IDHM_E', 'IDHM_L', 'IDHM_R']]
```

```python
mediaenemidhm_df=pd.merge(mediamuni_df, atlashumano_df, how='left', left_on=['CO_MUNICIPIO_RESIDENCIA'], right_on=['Codmun7'])
```

```python
mediaenemidhm_df.dtypes
```

```python
mediaenemidhm_df['IDHM'] = mediaenemidhm_df['IDHM'].apply(lambda x: float(str(x).replace(",",".")))
```

```python
mediaenemidhm_df.dtypes
```

```python
mediaenemidhm_df.plot.scatter('NU_NOTA_REDACAO', 'IDHM', figsize=(10,10))
```

```python
 mediaenemidhm_df.plot.scatter('NU_NOTA_MT', 'IDHM', figsize=(10,10))  
```

### Scatter pra nota de Redacao

```python
#variáveis
meanRedacao = mediaenemidhm_df.NU_NOTA_REDACAO.mean()
meanIDHM = mediaenemidhm_df.IDHM.mean()
idhmMunicipio=mediaenemidhm_df.IDHM
notaMunicipio=mediaenemidhm_df.NU_NOTA_REDACAO

#definindo a regra dos quadrantes
quadrante1 = (idhmMunicipio < meanIDHM) & (notaMunicipio < meanRedacao)
quadrante2 = (idhmMunicipio > meanIDHM) & (notaMunicipio < meanRedacao)
quadrante3 = (idhmMunicipio > meanIDHM) & (notaMunicipio > meanRedacao)
quadrante4 = (idhmMunicipio < meanIDHM) & (notaMunicipio > meanRedacao)

#filtrando os dataframes para criar os gráficos com os diferentes quadrantes
df1 = mediaenemidhm_df[quadrante1]
df2 = mediaenemidhm_df[quadrante2]
df3 = mediaenemidhm_df[quadrante3]
df4 = mediaenemidhm_df[quadrante4]

fig, ax = plt.subplots(2,2, figsize=(15,7), sharex='col', sharey='row')
plt.subplots_adjust(wspace=0, hspace=0)

minRedacao = mediaenemidhm_df.NU_NOTA_REDACAO.min()
maxRedacao = mediaenemidhm_df.NU_NOTA_REDACAO.max()
minIDHM = mediaenemidhm_df.IDHM.min()
maxIDHM = mediaenemidhm_df.IDHM.max()

df1.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=5, color='purple', ax = ax[1,0], xlim=(minRedacao, meanRedacao), ylim=(minIDHM, meanIDHM))
print("quadrante1: {} %, baixo IDM, baixa nota".format(100*len(df1.IDHM)/len(idhmMunicipio)))

df2.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=5, color='blue', ax = ax[0,0], xlim=(minRedacao, meanRedacao), ylim=(meanIDHM, maxIDHM))
print("quadrante1: {} %, alto IDM, baixa nota".format(100*len(df2.IDHM)/len(idhmMunicipio)))

df3.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=5, color='green', ax = ax[0,1], xlim=(meanRedacao, maxRedacao), ylim=(meanIDHM, maxIDHM))
print("quadrante1: {} %, alto IDM, alta nota".format(100*len(df3.IDHM)/len(idhmMunicipio)))

df4.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=5, color='yellow', ax = ax[1,1], xlim=(meanRedacao, maxRedacao), ylim=(minIDHM, meanIDHM))
print("quadrante1: {} %, baixo IDM, alta nota".format(100*len(df4.IDHM)/len(idhmMunicipio)))

```

### Scatter pra nota de matemática

```python
#variáveis
meanMat = mediaenemidhm_df.NU_NOTA_MT.mean()
meanIDHM = mediaenemidhm_df.IDHM.mean()
idhmMunicipio=mediaenemidhm_df.IDHM
notaMunicipio=mediaenemidhm_df.NU_NOTA_MT

#definindo a regra dos quadrantes
quadrante1 = (idhmMunicipio < meanIDHM) & (notaMunicipio < meanMat)
quadrante2 = (idhmMunicipio > meanIDHM) & (notaMunicipio < meanMat)
quadrante3 = (idhmMunicipio > meanIDHM) & (notaMunicipio > meanMat)
quadrante4 = (idhmMunicipio < meanIDHM) & (notaMunicipio > meanMat)

#filtrando os dataframes para criar os gráficos com os diferentes quadrantes
df1 = mediaenemidhm_df[quadrante1]
df2 = mediaenemidhm_df[quadrante2]
df3 = mediaenemidhm_df[quadrante3]
df4 = mediaenemidhm_df[quadrante4]

fig, ax = plt.subplots(2,2, figsize=(15,7), sharex='col', sharey='row')
plt.subplots_adjust(wspace=0, hspace=0)

minMat = mediaenemidhm_df.NU_NOTA_MT.min()
maxMat = mediaenemidhm_df.NU_NOTA_MT.max()
minIDHM = mediaenemidhm_df.IDHM.min()
maxIDHM = mediaenemidhm_df.IDHM.max()

df1.plot.scatter('NU_NOTA_MT', 'IDHM', s=5, color='orchid', ax = ax[1,0], xlim=(minMat, meanMat), ylim=(minIDHM, meanIDHM))
print("quadrante1: {} %, baixo IDM, baixa nota".format(100*len(df1.IDHM)/len(idhmMunicipio)))

df2.plot.scatter('NU_NOTA_MT', 'IDHM', s=5, color='silver', ax = ax[0,0], xlim=(minMat, meanMat), ylim=(meanIDHM, maxIDHM))
print("quadrante1: {} %, alto IDM, baixa nota".format(100*len(df2.IDHM)/len(idhmMunicipio)))

df3.plot.scatter('NU_NOTA_MT', 'IDHM', s=5, color='mediumpurple', ax = ax[0,1], xlim=(meanMat, maxMat), ylim=(meanIDHM, maxIDHM))
print("quadrante1: {} %, alto IDM, alta nota".format(100*len(df3.IDHM)/len(idhmMunicipio)))

df4.plot.scatter('NU_NOTA_MT', 'IDHM', s=5, color='lightblue', ax = ax[1,1], xlim=(meanMat, maxMat), ylim=(minIDHM, meanIDHM))
print("quadrante1: {} %, baixo IDM, alta nota".format(100*len(df4.IDHM)/len(idhmMunicipio)))
```

# Dispersão com latitude e longitude

```python
municipio_df = pd.read_csv("atlas.csv", sep=",")
```

```python
municipio_df=municipio_df[['Codmun7', 'Município', 'LONG', 'LAT', 'Alt']]
```

```python
enem_idhm_muni_df=pd.merge(mediaenemidhm_df, municipio_df, how='left', left_on=['CO_MUNICIPIO_RESIDENCIA'], right_on=['Codmun7'])
```

```python
enem_idhm_muni_df
```

```python
enem_idhm_muni_df.plot.scatter(y='LAT', x='LONG', 
                               c= enem_idhm_muni_df['NU_NOTA_REDACAO'],
                               cmap= 'RdYlGn',
                               title='NU_NOTA_REDACAO',
                               figsize=(10,9))
```
